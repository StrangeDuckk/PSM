Gra w życie (Life, The game of life) – jeden z pierwszych i najbardziej znanych przykładów automatu
komórkowego, wymyślony w roku 1970 przez brytyjskiego matematyka Johna Conwaya.
Gra toczy się na nieskończonej planszy (płaszczyźnie) podzielonej na kwadratowe komórki. Każda
komórka ma ośmiu „sąsiadów” (tzw. sąsiedztwo Moore’a), czyli komórki przylegające do niej bokami
i rogami. Każda komórka może znajdować się w jednym z dwóch stanów: może być albo „żywa”
(włączona), albo „martwa” (wyłączona). Stany komórek zmieniają się w pewnych jednostkach czasu.
Stan wszystkich komórek w pewnej jednostce czasu jest używany do obliczenia stanu wszystkich
komórek w następnej jednostce. Po obliczeniu wszystkie komórki zmieniają swój stan dokładnie w
tym samym momencie. Stan komórki zależy tylko od liczby jej żywych sąsiadów. W grze w życie nie
ma graczy w dosłownym tego słowa znaczeniu. Udział człowieka sprowadza się jedynie do ustalenia
stanu początkowego komórek.


Reguły gry według Conwaya
• Martwa komórka, która ma dokładnie 3 żywych sąsiadów, staje się żywa w następnej
jednostce czasu (rodzi się)

• Żywa komórka z 2 albo 3 żywymi sąsiadami pozostaje nadal żywa; przy innej liczbie
sąsiadów umiera (z „samotności” albo „zatłoczenia”).

Reguły, jakim podlega automat opisywane są często skrótowo w następujący sposób: •

• następnie umieszcza się ukośnik: /;

• po ukośniku umieszcza się te liczby komórek w sąsiedztwie, dla których martwe

Zadanie
Zaimplementować automat komórkowy działający na skończonej planszy z cyklicznymi warunkami
brzegowymi, z możliwością zadania reguł gry

stałe i importy

```python
import numpy as np

wielkoscMapy = int(input("podaj wielkosc mapy nxn: "))
zasadaPrzezycie,zasadaNarodziny = input("podaj zasady przezycie/smierc: ").split("/")
zasadaPrzezycie = set(int(a) for a in zasadaPrzezycie) # {2,3} przezycie przy dwoch i trzech komorkach
zasadaNarodziny = set(int(a) for a in zasadaNarodziny) # {3} narodziny przy 3 komorkach
plansza = np.zeros((wielkoscMapy,wielkoscMapy),dtype=int) #tablica dwuwymiarowa kwadratowa
```

wypisanie planszy w formacie □ i ■

```python
def wypisaniePlanszy():
    for i in range(wielkoscMapy):
        for j in range(wielkoscMapy):
            if(plansza[i][j]==1):
                print("  ■",end = "")
            else:
                print("  □", end = "")
        print()
```

index lewej gornej komorki zaczyna

```python
indexI = 1 #int(input("podaj umiejscowienie glidera w chwili 0 (i): ");
indexJ = 1 #int(input("podaj umiejscowienie glidera w chwili 0 (j): ");
plansza[indexI][indexJ+2] = 1
plansza[indexI+1][indexJ] = 1
plansza[indexI+1][indexJ+2] = 1
plansza[indexI+2][indexJ+1] = 1
plansza[indexI+2][indexJ+2] = 1

wypisaniePlanszy()
```

obliczanie sasiadow z cyklicznymi brzegami

```python
def obliczSasiadow(x,y):
    suma = 0
    for i in [-1,0,1]: #w szerz sa 3 komorki
        for j in [-1,0,1]: #w zdluz tez sa 3 komorki
            if i == 0 and j == 0: #jestesmy w srodkowej komorce
                continue
            #dla obliczenia brzegow cyklicznych
            nowyX = (x+i) % wielkoscMapy
            nowyY = (y+j) % wielkoscMapy
            suma += plansza[nowyX][nowyY]
    return suma
```

pojedynczy krok

```python
def krok():
    newPlansza = np.zeros_like(plansza)
    for i in range(wielkoscMapy):
        for j in range(wielkoscMapy):
            sasiedziKomorki = obliczSasiadow(i,j)
            if(plansza[i][j]==1) and sasiedziKomorki in zasadaPrzezycie:
                newPlansza[i][j] = 1
            elif (plansza[i][j] == 0) and sasiedziKomorki in zasadaNarodziny:
                newPlansza[i][j] = 1
    return newPlansza
```

```python
licznik = 1
print("krok: " , 0)
wypisaniePlanszy()
while True:
    print("\nkrok: " , licznik)
    licznik += 1
    plansza = krok()
    wypisaniePlanszy()
    ui = input("enter aby przejsc do kolejnego kroku albo q (jupiter = cancel) aby zakonczyc")
    if(ui == "q"):
        break
```
